## Web scrapping de IMDB

EJERCICIO: DESCARGA UN DATAFRAME CON EL TOP 250 PELÍCULAS DE IMBD A TRAVÉS DE WEBSCRAPING.

Obtén:
* Título
* Año
* Posición
* Reparto
* Rating
* Votos

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# Download IMDB's Top 250 data
url = 'http://www.imdb.com/chart/top'
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
title = soup.find("td", class_ = "titleColumn").get_text(strip=True).split('.')[1][:-6]
title
```

```python
año = int(soup.find("span", class_ = "secondaryInfo").get_text()[1:-1])
año
```

```python
title = int(soup.find("td", class_ = "titleColumn").get_text(strip=True).split('.')[0])
title
```

```python
rating = float(soup.find("td", class_="ratingColumn imdbRating").get_text(strip=True))
rating
```

```python
crew = soup.find("td", class_ = "titleColumn").find('a').get("title")
crew
```

```python
votes = int(soup.find("td", class_="posterColumn").find_all("span")[3].get('data-value'))
votes
```

```python
movies = soup.find_all('td', class_="titleColumn")
movies
```

```python
titles = [x.get_text(strip=True).split('.')[1][:-6] for x in movies]
titles[:5]
```

```python
years = [int(x.get_text(strip=True)[-5:-1]) for x in movies]
print(years[:5])
crews = [x.find('a').get('title') for x in movies]
print(crews[:5])
```

```python
movies_val = soup.find_all("td", class_="posterColumn")
movies_val
```

```python
rankings = [int(x.find("span").get('data-value')) for x in movies_val]
print(rankings[0:5])
ratings = [round(float(x.find_all("span")[1].get('data-value')),2) for x in movies_val]
print(ratings[0:5])
votes = [int(x.find_all("span")[3].get('data-value')) for x in movies_val]
print(votes[0:5])
```

```python
pd.set_option("max_colwidth", 100)
```

```python
dict_movies = {
    "título": titles,
    "años": years,
    "reparto": crews,
    "ranking": rankings,
    "valoración": ratings,
    "votaciones": votes
}
results = pd.DataFrame(dict_movies)
results.head(10)
```
