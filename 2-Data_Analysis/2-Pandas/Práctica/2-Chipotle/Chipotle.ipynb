### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipo = pd.read_csv(url, sep="\t")
chipo
```

```python
pd.read_table(url)
```

### Step 3. Name of the max valued product

```python
chipo['item_price'].max()
```

```python
# chipo['item_price'].str.replace("$", "")
```

```python
chipo['item_price'] = chipo['item_price'].str[1:].astype(float)
```

```python
chipo['item_price'].max()
```

```python
chipo.info()
```

```python
chipo[chipo['item_price'] == chipo['item_price'].max()]
```

```python
chipo['unit_price'] = chipo['item_price'] / chipo['quantity']
chipo
```

```python
lista_vp = list(chipo[chipo['unit_price'] == chipo['unit_price'].max()]['item_name'].unique())
lista_vp
```

```python
chipo_vp = chipo[chipo['item_name'].isin(lista_vp)]
chipo_vp
```

```python
chipo_vp.groupby('item_name')['item_price'].sum().idxmax()
```

### Step 4. How many products cost more than $10.00?

```python
len(chipo[chipo['unit_price'] > 10]['item_name'].unique())
```

Extra: Cuántos pedidos se han hecho con un producto de más de 10$
